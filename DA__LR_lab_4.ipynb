```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/TarekDib03/Analytics/master/Week2%20-%20Linear%20Regression/Data/climate_change.csv")
```

```python
df
```

```python
df.info()
```

```python
input = df.iloc[:,:10]
output = df.iloc[:,10]
output
```

```python
input
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

x_train,x_test,y_train,y_test = train_test_split(input,output,test_size = 0.2,random_state = 0)
```

```python
linreg=LinearRegression()
linreg.fit(x_train,y_train)
```

```python
y_pred=linreg.predict(x_test)
y_pred
```

```python
Accuracy=r2_score(y_test,y_pred)*100
print(" Accuracy of the model is %.2f" %Accuracy)
```
