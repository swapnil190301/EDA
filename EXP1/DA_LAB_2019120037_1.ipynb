**DATA ANALYTICS LAB EXPERIMENT 1: Exploratory Data Analysis**

**Objective**: Perform EDA such as number of data samples, number of features, number of classes, number of data samples per class, removing missing values, conversion to numbers, using seaborn library to plot different graphs.

```python
#importing all packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp
import seaborn as sb
```

```python
#reading the dataset
df = pd.read_csv("/content/DataPanelWHR2021C2.csv")
df.head()
```

```python
#knowing the dataset
df.tail()
```

```python
#knowing the dataset
df.info()
```

```python
#cleaning the dataset
df=df.dropna()
```

```python
#knowing the dataset
df.describe()
```

```python
#knowing the dataset
df.shape
```

```python
#grouping the countries by name
gk = df.groupby('Country name')
sc = lambda x: x.mean()
data = gk.transform(sc)
data['Country name']=df['Country name']
```

```python
titles=list(data.columns)
titles[0], titles[10] = titles[10], titles[0]
titles
data=data[titles]
```

```python
#giving a single entry for each country by their respective mean values
data=data.drop_duplicates()
data.head()
```

```python
#converting year into int because year can't be float
data['year']=data['year'].astype(int)
data.head()
```

```python
data.info()
```

```python
#Sorting according to positive affect
data.sort_values('Positive affect',ascending=False)
```

Iceland and Syria have the highest and lowest positive affect respectively.

```python
#Sorting according to negative affect
data.sort_values('Negative affect')
```



Creating a new attribute happiness parameter since positive affect and negatuve affect are giving different solutions and that isn't an optimal solution.

```python
data['happiness_parameter']=data['Positive affect']-data['Negative affect']
data=data.sort_values('happiness_parameter',ascending=False)
data.describe()
```

```python
data.head()
```

Iceland has the highest happiness parameter, so Iceland is the happiest country among all countries mentioned in the dataset.

```python
data.tail()
```

```python
#calculating correlation
data.corr()
```

```python
#plotting correlation using heatmap
dataplot = sb.heatmap(data.corr(), cmap="YlGnBu", annot=True)
mp.show()
```

```python
#removing unnecesary columns for further process
df = pd.DataFrame(data)
df.drop(df.columns[[0, 8, 9, 10]], axis=1, inplace=True)
df
```

```python
#calculating correlation with attributes and target variable
df.corrwith(df['happiness_parameter'])
```

```python
#plotting correlation with attributes and target variable
mp.plot(['LL', 'GDP', 'SS', 'LE', 'F', 'G', 'PC', 'HP'],df.corrwith(df['happiness_parameter']).values)
```

**OBSERVATION:**



*   Iceland and Syria have the highest and lowest positive affect respectively.
*Iraq and Taiwan Province of China have the highest and lowest negative affect respectively.
*   Iceland has the highest happiness parameter.
*   Iraq has the lowest happiness parameter.
